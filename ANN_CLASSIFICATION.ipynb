```python
import pandas as pd
```

```python
data=pd.read_csv('DateFruit_Dataset.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
data['Class']=lbl.fit_transform(data['Class'])
x=data.drop('Class',axis=1)
y=data['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scl=StandardScaler()
x_train_scaled=scl.fit_transform(x_train)
x_test_scaled=scl.fit_transform(x_test)
```

```python
import torch 
import torch.nn as nn
```

```python
x_train_tensor=torch.tensor(x_train_scaled,dtype=torch.float32)
y_train_tensor=torch.tensor(y_train.values,dtype=torch.long)
x_test_tensor=torch.tensor(x_test_scaled,dtype=torch.float32)
y_test_tensor=torch.tensor(y_test.values,dtype=torch.long)
```

```python
from torch.utils.data import TensorDataset,DataLoader
train_tensordataset=TensorDataset(x_train_tensor,y_train_tensor)
test_tensordataset=TensorDataset(x_test_tensor,y_test_tensor)
```

```python
train_dataloader=DataLoader(train_tensordataset,batch_size=32,shuffle=True)
test_dataloader=DataLoader(test_tensordataset,batch_size=32)
```

```python
class ANN_CLASSIFICATION(nn.Module):
    def __init__(self):
        super(ANN_CLASSIFICATION,self).__init__()
        self.model=nn.Sequential(
            nn.Linear(x.shape[1],64),
            nn.ReLU(),
            nn.Linear(64,64),
            nn.ReLU(),
            nn.Linear(64,7)
        )
    def forward(self,x):
        return self.model(x)
```

```python
model=ANN_CLASSIFICATION()
loss_criteria=nn.CrossEntropyLoss()
import torch.optim as optim
optimizer=optim.Adam(model.parameters())
```

```python
epochs=101

for epoch in range(epochs):
    model.train()
    loss_train=0.0
    for xb,yb in train_dataloader:
        optimizer.zero_grad()
        output=model(xb)
        loss=loss_criteria(output,yb)
        loss.backward()
        optimizer.step()
        loss_train+=loss.item()
    val=loss_train/len(train_dataloader)
    print(f'epochs{epoch} training_loss{val}')
    
        
```

```python
model.eval()
tot=0
pred=0
with torch.no_grad():
    for xb,yb in test_dataloader:
        output=model(xb)
        v,i=torch.max(output,1)
        pred+=(i==yb).sum().item()
        tot+=yb.size(0)
        
        
```
